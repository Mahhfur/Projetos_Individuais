```python
#calcula seno de x
def seno(x):
  a=x*np.ones(30)
  exp = np.arange(1,60,2)
  a=a**exp
  b=factorial(exp) #o expoente eh o mesmo número do denominador

  c= a/b
  c[1::2] = -1*c[1::2]
  mysin = c.sum()
  print (mysin)

```

```python
#calcula cosseno de x
def cos(x):
  a=x*np.ones(30)
  exp = np.arange(0,59,2)
  a=a**exp
  b=factorial(exp)

  c= a/b
  c[1::2] = -1*c[1::2]
  mycos = c.sum()
  print (mycos)
```

```python
# método de Monte Carlo para estimar a área abaixo do gráfico da função y=x^2 no intervalo [0,1]. 
def areax2mc(k):
  x=np.random.rand(k)
  y=np.random.rand(k)
  xx= np.linspace(0,1,100)
  yy=(xx**2)
  plt.scatter(x,y)
  plt.plot(xx,yy,'black')
  plt.figure()

  plt.plot(xx,yy,'black')
  sel = y < x**2 
  plt.scatter (x[sel], y[sel])
  plt.figure()
  
  area= sel.sum()/k
  return area


areax2mc(100)
```

```python
#gráfico da função seno(x) no intervalo de -2π a 2π.

x= np.linspace(-2*np.pi, 2*np.pi ,150)
y=np.sin(x)
plt.plot(x,y)
plt.title('função seno(x)')

plt.show()
```

```python
#gráfico da função cosseno(x) no intervalo de -2π a 2π

x= np.linspace(-2*np.pi, 2*np.pi ,150)
y=np.cos(x)
plt.plot(x,y)
plt.title('função cos(x)')
plt.fill_between(x,y)

plt.show()
```

```python
#gráfico da função tangente(x) no intervalo de -2π a 2π

x= np.linspace(-2*np.pi, 2*np.pi ,150)
y = np.tan(x) 
plt.scatter(x,y)
plt.title('função tg(x)')
plt.xlabel('valor de x')

plt.show()
```
