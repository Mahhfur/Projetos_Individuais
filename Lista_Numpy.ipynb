EXERCICÍO 4 

Crie uma função na qual calcula o valor do cos a partir da série de Taylor
(50 primeiros termos) e seno a partir da seguinte identidade. Obs: Fazer
a serie sem utilizar for/ while. Utilizar a função factorial da biblioteca scipy
(https://docs.scipy.org/doc/scipy/reference/generated/scipy.special.facto
rial.html).

```python
from scipy.special import factorial
import numpy as np
from math import pi

def Cos(x):
    n = np.arange(0, 51, 1)
    signal = pow(-1, n)
    x = pow((x*np.ones(n.size)), (2*n))
    num = signal * x
    den = factorial(2*n)
    cos = (num/den).sum()
    return cos

def Sen(x):
    n = np.arange(0, 51, 1)
    x = pow((x*np.ones(n.size)), (2*n + 1))
    signal = pow(-1, n)
    num = signal * x
    den = factorial(2*n + 1)
    sen = (num/den).sum()
    return sen


x = float(input("Digite x: "))
cos = Cos(x)
sen = Sen(x)
print(f'cos(x) = {cos:.3f}')
print(f'sen(x) = {sen:.3f}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Cria um array 'x' que varia de 0 a pi, com 200 pontos
# e calcula os valores de seno correspondentes em 'y'
x = np.linspace(0, np.pi, 200)
y = np.sin(x)

# Plota a curva de seno usando os valores calculados em 'x' e 'y'
plt.plot(x, y)

# Gera 3 milhões de pontos aleatórios no retângulo que circunscreve o gráfico do seno
xx = np.pi * np.random.rand(3000000)
yy = np.random.rand(3000000)

# Cria um gráfico de dispersão desses pontos aleatórios
plt.scatter(xx, yy)

# Seleciona os pontos que caem abaixo da curva de seno
# e cria um novo gráfico de dispersão mostrando apenas esses pontos
sel = yy < np.sin(xx)
plt.figure()
plt.plot(x, y)
plt.scatter(xx[sel], yy[sel])

# Calcula a área da região sob a curva de seno
# usando a contagem de pontos selecionados e o tamanho total da amostra aleatória
area = np.pi * sel.sum() / 3000000

# Imprime o valor da área
print(area)
```
